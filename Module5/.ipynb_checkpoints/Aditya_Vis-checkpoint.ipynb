```python
def drawLine(df,channel, title):
    df_filter = df[df['Channel']==channel]
    x=len(df_filter)
    date_label=df_filter.Date.unique()
    fig = plt.figure(figsize=(12,6))
    ax = fig.add_subplot(111,facecolor='lavenderblush')
    #ax.scatter(X_test, y_test, c='g', marker='o')
    ax.plot(df_filter.Date,df_filter['Actual Sales'], color='blue', linewidth=2, alpha=0.7)
    ax.plot(df_filter.Date, df_filter['Challenger Forecast'], color='red', linewidth=2, alpha=0.7)
    ax.plot(df_filter.Date, df_filter['Champion Forecast'], color='black', linewidth=2, alpha=0.7)
    for i,j in zip(df_filter.Date,df_filter['Actual Sales']):
        ax.annotate(str(j),xy=(i,j))
    ax.set_title(title,fontsize=20,weight='bold')
    #ax.xaxis.set_major_locator(plt.MaxNLocator(8))
    #ax.set_xticks(ax.get_xticks()[::1])
    ax.set_xlabel('Date',fontsize=10,weight='bold')
    ax.set_ylabel('Channel',fontsize=10,weight='bold')
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, labels)
    print(title)
    plt.show()
```

```python
%matplotlib inline
c =['Channel A','Channel B','Channel C']
for i in range(0,len(c)):
    title = c[i]+' Performance'
    drawLine(df,c[i],title)
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
# data to plot
n_groups = 4
means_frank = (90, 55, 40, 65)
means_guido = (85, 62, 54, 20)
 
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
rects1 = plt.bar(index, means_frank, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Frank')
 
rects2 = plt.bar(index + bar_width, means_guido, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Guido')
 
plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D'))
plt.legend()
 
plt.tight_layout()
plt.show()
```

```python

def auto_plot_bar(df,channel,title):
    df_filter = df[df['Channel']==channel]
    index = df_filter['Date'].unique()
    #fig,ax = plt.subplots(figsize=(20,6),facecolor='lavenderblush')
    fig = plt.figure(figsize=(20,6))
    ax = fig.add_subplot(111,facecolor='white')
    y_pos = np.arange(len(index))
    bar_width = 0.35
    opacity = 0.4
    plt.grid(zorder=0)
    plt.bar(y_pos, df_filter['Challenger Difference'], bar_width,alpha=opacity,color='red',label='ActualSales-ChallengerForecast')
    for i,j in zip(y_pos,df_filter['Challenger Difference']):
        if j>0:
            ax.annotate(str(j),xy=(i-0.1,j+10),color='blue')
        else:
            ax.annotate(str(j),xy=(i-0.15,j-45),color='red')
    plt.bar(y_pos+bar_width, df_filter['Champion Difference'], bar_width,alpha=opacity,color='orange',label='ActualSales-ChampionForecast')
    for i,j in zip(y_pos+bar_width, df_filter['Champion Difference']):
        if j>0:
            ax.annotate(str(j),xy=(i-0.1,j+10),color='blue')
        else:
            ax.annotate(str(j),xy=(i-0.15,j-45),color='red')
    plt.title(title+'\n',fontsize=25,weight='bold')
    plt.xticks(y_pos+bar_width/2, index)
    plt.tight_layout()
    plt.legend()
    plt.show()
```

```python
%matplotlib inline
c =['Channel A','Channel B','Channel C']
for i in range(0,len(c)):
    title = c[i]+' Compare Difference between ActualSales/Challenger Vs ActualSales/Champion'
    auto_plot_bar(df,c[i],title)
```
