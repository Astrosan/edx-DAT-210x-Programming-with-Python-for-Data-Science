```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# This is new
from pandas.tools.plotting import parallel_coordinates
```

```python
# Look pretty...

# matplotlib.style.use('ggplot')
plt.style.use('ggplot')
```

Load up the wheat seeds dataset into a dataframe. We've stored a copy in the Datasets directory.

```python
# .. your code here ..
wheat=pd.read_csv('Datasets/wheat.data',index_col=0)
```

If you loaded the `id` column as a feature (hint: _you shouldn't have!_), then be sure to drop it. Also get rid of the `area` and `perimeter` features:

```python
# .. your code here ..
wheat.drop(columns=['area','perimeter'],inplace=True)
wheat
```

Plot a parallel coordinates chart grouped by the `wheat_type` feature. Set the optional display parameter `alpha` to `0.4`:

```python
# .. your code here ..
plt.style.use('seaborn')
from pandas.tools.plotting import parallel_coordinates
plt.figure(figsize=(10,6))
parallel_coordinates(wheat,'wheat_type',alpha=0.4, colormap=plt.get_cmap("Set1"))
plt.show()
```

Create a 2d scatter plot that graphs the `compactness` and `width` features:

```python
# Display the graphs:
plt.show()
wheat.plot.scatter(x='compactness',y='width',color='r',marker='^',alpha=0.5)
plt.show()
```

```python
from sklearn.datasets import load_iris
from pandas.tools.plotting import parallel_coordinates
# Look pretty...
matplotlib.style.use('ggplot')
# If the above line throws an error, use plt.style.use('ggplot') instead
	
# Load up SKLearn's Iris Dataset into a Pandas Dataframe
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names) 
	
df['target_names'] = [data.target_names[i] for i in data.target]
	
# Parallel Coordinates Start Here:
plt.figure()
parallel_coordinates(df, 'target_names')
plt.show()
```
